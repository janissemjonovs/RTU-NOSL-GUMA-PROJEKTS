# LNSO (Latvijas Nacionālais simfoniskais orķestris) publiski pieejamo finanšu datu apstrāde. 2016.-2019.gads.#

### https://www.lnso.lv/darbibas-parskats 

```python
data = { "Gads" : [2016, 2017, 2018, 2019],
            "Jauno programmu skaits" : [49, 51, 52, 40],
            "Koncerti bērniem un jauniešiem" : [20, 25, 37, 100],
            "Koncertu skaits kopā" : [60, 82, 80, 61 ],
            "Apmeklētāju skaits": [45114, 56025, 47450, 38502],
            "Zāles piepildījums(%)" : [78, 70, 74, 75],
            "Biļešu ieņēmumi": [329081, 252374, 289600, 309710],
            "Pārējie ieņemumi": [296873, 438260, 354400,  329222]
            
    
}

print(data)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df = df.set_index("Gads")  # Gads = indekss
```

```python
df
```

```python
df["Koncertu skaits kopā"].plot()
df["Koncerti bērniem un jauniešiem"].plot()
df["Jauno programmu skaits"].plot()
df["Zāles piepildījums(%)"].plot()
plt.title("Lnso datu analīze")
plt.legend(["Visi koncerti", "Jauniešu koncerti", "Jaunās programmas", "Zāles piepildījums"])
```

## Grafiski atainot "Biļešu ieņēmumus" un "Pārējos ieņēmumus" ##

```python
df["Biļešu ieņēmumi"].plot()
df["Pārējie ieņemumi"].plot()
plt.title("Ieņēmumu tendences")
plt.legend(["Biļešu ieņēmumi", "Pārējie ieņēmumi"])
```

## Kāds ir bijis vismazākais apmeklētāju skaits?##

### Atainot "Biļešu ieņēmumus" grafikā, stabiņos 

```python

x = [2016,2017,2018,2019]   # parādīt biļešu ieņēmumus
y = [329081, 252374, 289600, 309710]
plt.bar(x,y)
plt.title("Biļešu ieņēmumi")
plt.show()
```

### Atainot tendences, izmantojot Stacklpot vizualizāciju, jomās:  "Biļešu ieņēmumi", "Pārējie ieņēmumi", "Apmeklētāju skaits".

```python

x = [2016,2017,2018,2019]
y1 = [329081, 252374, 289600, 309710]  # Biļešu iņēmumi,
y2 = [296873, 438260, 354400, 329222]  # Pārejie ieņēmumi
y3 = [45114, 65025, 47450, 38502] # Apmeklētāju skaits
plt.stackplot(x, y1, y2, y3)
plt.show()
```

## Izmantojot rīku "Pie chart", parādīt proporciju starp 2107.gada pozīcijām "Biļešu ieņēmumi" un "Pārējie ieņēmumi", un 2019.gada tām pašām pozīcijām.

```python
labels = ["Biļešu ieņēmumi2017", "Pārējie ieņēmumi2017"]
plt.pie([252374, 438260], labels=labels, autopct="%1.0f%%")
plt.show()

labels = ["Biļešu ieņēmumi2019", "Pārējie ieņēmumi2019"]
plt.pie([309710, 329222], labels=labels, autopct="%1.0f%%")
plt.show()

```
